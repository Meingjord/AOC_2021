```python
def parse(input):
    result = {}
    row = 0
    mapping = {}
    header = input[0]
    for i in range(len(header)):        
        mapping[i] = 1 if header[i] == "#" else 0

    for line in input[2:]:
        col = 0
        for c in line:
            point = (col,row)
            if c == "#":
                result[point] = 1
            col += 1
        row += 1
    return (result, mapping)

def print_field(input):
    x_low = 0
    x_high = 0    
    y_low = 0
    y_high = 0
    for point in input:
        x_low = min(x_low, point[0])
        x_high = max(x_high, point[0])
        y_low = min(y_low, point[1])
        y_high = max(y_high, point[1])
    for y in range(y_low, y_high+1):
        line = ''
        for x in range(x_low, x_high+1):        
            point = (x,y)
            if point in input:
                line += str(input[(x,y)])
            else:
                line += ' '
        print(line)
    print('')

print_field(parse(lines_example1)[0])

parse(lines)[1]
```

```python
def to_field_code(point, input):
    result = 0
    for y in range(-1,2):
        for x in range(-1,2):
            result *= 2
            shifted = (x + point[0],y+point[1])
            if shifted in input:
                result += 1
    return result   

def iterate(input, mapping):
    result = {}
    x_low = 0
    x_high = 0    
    y_low = 0
    y_high = 0
    for point in input:
        x_low = min(x_low, point[0])
        x_high = max(x_high, point[0])
        y_low = min(y_low, point[1])
        y_high = max(y_high, point[1])
    x_low -= 60
    x_high += 60
    y_low -= 60
    y_high += 60
    for x in range(x_low, x_high+1):
        for y in range(y_low, y_high+1):
            point = (x,y)
            index = to_field_code(point, input)
            if mapping[index] > 0:
                result[point] = 1
    return result

def count_field(data, input, i):
    x_low = 0
    x_high = 0    
    y_low = 0
    y_high = 0
    for point in input:
        x_low = min(x_low, point[0])
        x_high = max(x_high, point[0])
        y_low = min(y_low, point[1])
        y_high = max(y_high, point[1])
    x_low -= i
    y_low -= i
    x_high += i
    y_high += i
    total = 0
    for y in range(y_low, y_high+1):
        for x in range(x_low, x_high+1):        
            point = (x,y)
            if point in data:
                total += 1
    return total

def clean_border(data, input, i):
    x_low = 0
    x_high = 0    
    y_low = 0
    y_high = 0
    for point in input:
        x_low = min(x_low, point[0])
        x_high = max(x_high, point[0])
        y_low = min(y_low, point[1])
        y_high = max(y_high, point[1])
    x_low -= i
    y_low -= i
    x_high += i
    y_high += i
    result = {}
    for y in range(y_low, y_high+1):
        for x in range(x_low, x_high+1):        
            point = (x,y)
            if point in data:
                result[point] = data[point]
    return result    

def calculate(input, n, debug):
    data, mapping = input
    current = data
    for i in range(n):
        print("step", i)
        current = iterate(current, mapping)
        if i % 2 == 1:
            current = clean_border(current, data, i+1)
        #else:
            #current = grow_border(current, data, i+1)

        if debug:
            print()
            print_field(current)
    print(count_field(current, data, n))

def calculate_part_one(input):
    calculate(input, 2, True)
    
calculate_part_one(parse(lines_example1))
```
