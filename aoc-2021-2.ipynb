```python
import pandas as pd
import os
import unittest
import string
import re
```

```python
trainFile = "C:/Code/advent/2021_day2_input.txt"
pwd = os.getcwd()
os.chdir(os.path.dirname(trainFile))
file1 = open(trainFile, 'r') 
lines = file1.read().splitlines()
#s = list(map(string, lines[0].split(",")))
#series = pd.Series(s)
```

```python
tuples = list(map(lambda x: x.split(" "),lines))
tuples
```

```python
# PART 1
depth = 0
horizontal = 0
for command in tuples:
    verb = command[0]
    value = int(command[1])
    if verb == "forward":
        horizontal += value
    elif verb == "down":
        depth += value
    elif verb == "up":
        depth -= value
(depth,horizontal,depth*horizontal)
    
```

```python
# PART 2
aim = 0
horizontal = 0
depth = 0
for command in tuples:
    verb = command[0]
    value = int(command[1])
    if verb == "forward":
        horizontal += value
        depth += aim * value
    elif verb == "down":
        aim += value
    elif verb == "up":
        aim -= value
(depth,horizontal,depth*horizontal)
```
