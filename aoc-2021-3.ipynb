```python
def calculate_part1(input):
    n = len(input)
    m = len(input[0])
    transposed = list(map(list, zip(*input)))
    counts = [sum(x) for x in transposed]
    gamma = [1 if 2*x>n else 0 for x in counts]
    epsilon = [1 -x for x in gamma]
    return (gamma,epsilon)
```

```python
def vector_to_int(v):
    count = len(v)
    e = 0
    f = 1
    for i in range(count):
        e += f if v[count-i-1] else 0
        f *= 2
    return e
```

```python
def multiply_vectors(v):
    (v1, v2) = v
    count = len(v1)
    e = 0
    g = 0
    f = 1
    for i in range(count):
        e += f if v1[count-i-1] else 0
        g += f if v2[count-i-1] else 0
        f *= 2
    return (e,g,e*g)
```

```python
def calculate_part2_oxygen(input):
    data = copy.deepcopy(input)
    pos = 0
    while (len(data)> 1):
        n = len(data)
        indexes = list(map(lambda x: x[pos], data))
        #print(indexes)
        s = sum(indexes)
        p = s*2 >= n
        data = list(filter(lambda x: x[pos] == p, data))
        pos += 1
    print(len(data), vector_to_int(data[0]))
    return data[0]
```

```python
calculate_part2_oxygen(exampleTuples)
```

```python
def calculate_part2_co2(input):
    data = copy.deepcopy(input)
    pos = 0
    while (len(data)> 1):
        #print('scanning pos' , pos , data)
        n = len(data)
        indexes = list(map(lambda x: x[pos], data))
        #print(indexes)
        s = sum(indexes)
        p = 0 if s*2 >= n else 1
        #print ('looking for',p)
        data = list(filter(lambda x: x[pos] == p, data))
        pos += 1
        #print ('new data', pos)
        #print(data)
    print(len(data), vector_to_int(data[0]))
    return data[0]
```

```python
calculate_part2_co2(exampleTuples)
```

```python
def calculate_part2(input):
    o2 = calculate_part2_oxygen(input)
    co2 = calculate_part2_co2(input)
    print(multiply_vectors((o2, co2)))
```
