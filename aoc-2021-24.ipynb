```python
to_register = {'w':0,'x':1, 'y':2, 'z':3}

def parse_line(input, location):
    parts = input.split(' ')
    verb = parts[0]
    if verb == 'inp':
        location[0] = location[0]+1
        return [verb,parts[1],to_register[parts[1]],None, location[0]]
    
    if parts[2] in to_register:
        return [verb,parts[1],to_register[parts[1]],parts[2],to_register[parts[2]]]
    else:
        return [verb,parts[1],to_register[parts[1]],None,int(parts[2])]

def parse(input):
    result = []
    location = [-1]
    for line in input:
        result.append(parse_line(line,location))
    return result

parse(lines_example1)
```

```python
def do_operator(operator, left, right):
    if operator == 'add':        
        return left + right
    if operator == 'mul':
        return left * right
    if operator == 'div':
        return left // right
    if operator == 'mod':
        return left % right
    if operator == 'eql':
        return 1 if left == right else 0
    
    return None

def do_instruction(registers, instruction, data):    
    verb = instruction[0]
    if verb == 'inp':
        registers[instruction[2]] = data[instruction[4]]
    elif instruction[3] == None:
        registers[instruction[2]] = do_operator(verb, registers[instruction[2]], instruction[4])
    else:
        registers[instruction[2]] = do_operator(verb, registers[instruction[2]],registers[instruction[4]])    
    print(instruction, registers)

def evaluate(instructions, data):
    registers = [0,0,0,0]
    for instruction in instructions:
        do_instruction(registers, instruction, data)    
    return registers

def string_to_data(input):
    return list(map(int, input))

print(string_to_data('13579246899999'))
evaluate(parse(lines_example1),string_to_data('9'))
```

```python
def to_magic_numbers(input):
    result = []
    n = len(input) // 14
    for i in range(0,14):
        index = i * n
        magic = (input[4+ index][4], input[5+ index][4], input[15+ index][4])
        result.append(magic)

    return result

problem_magic_numbers = to_magic_numbers(parse(lines))
problem_magic_numbers
```

```python
def do_magic(registers, magic, data):
    registers[0] = data
    registers[1] = registers[3] % 26
    registers[3] = registers[3] // magic[0]
    registers[1] = 0 if registers[1] + magic[1] == registers[0] else 1
    registers[2] = (25 * registers[1]) + 1
    registers[3] = registers[3] * registers[2]
    registers[2] = (registers[0] + magic[2]) * registers[1]
    registers[3] = registers[3] + registers[2]

def evaluate_magic(instructions, data):
    registers = [0,0,0,0]
    magic = to_magic_numbers(instructions)
    position = 0
    for m in magic:
        do_magic(registers, m, data[position])
        position += 1
    return registers

evaluate_magic(parse(lines),string_to_data('12345678912345'))
```

```python
def do_magic2(z, magic, input):
    x = 0 if z % 26 + magic[1] == input else 1
    y1 = (25 * x) + 1
    z = (z // magic[0])* y1 + (input + magic[2]) * x
    return z

def do_magic3(z, magic, input):
    z_divided = z // magic[0]
    if z % 26 + magic[1] == input:
        z = z_divided
    else:
        z = z_divided * 26 + input + magic[2]
    return z

def evaluate_magic2(instructions, data):
    magic = to_magic_numbers(instructions)
    position = 0
    carry = 0
    for m in magic:
        carry = do_magic3(carry, m, data[position])
        position += 1
    return carry

def do_magic3_debug(z, magic, input):
    print('------------------------------------------')
    print('checking', z, magic, input)
    z_divided = z // magic[0]
    print('condition',z % 26 + magic[1], input, z % 26 + magic[1] - input )
    if z % 26 + magic[1] == input:
        print('divide by', magic[0])
        print('new', z_divided)
        return z_divided
    else:
        if magic[0] == 1:
            print('grow', magic[0], z_divided, input, magic[2])
        else:
            print('add', magic[0], z_divided, input, magic[2])
        print('new', z_divided * 26 + input + magic[2])
        return z_divided * 26 + input + magic[2]

def evaluate_magic2_debug(instructions, data):
    magic = to_magic_numbers(instructions)
    position = 0
    carry = 0
    for m in magic:
        carry = do_magic3_debug(carry, m, data[position])
        position += 1
    return carry

#evaluate_magic2_debug(parse(lines),string_to_data('12345678912345'))
```
